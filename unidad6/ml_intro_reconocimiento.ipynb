Tenemos una serie de variables de entrada utilizadas para entrenar un algoritmo de machine learning, esperando una salida o un objetivo. Ejemplo, trading. Podemos utilizar diferentes variables de entrada como: precios anteriores de la misma acción, precios de otros indicadores económicos, noticias, estados de ánimo de la población, etc. Todos esos datos van a entrenar este modelo y, oportunamente, podría ayudarnos a predecir el precio futuro de la acción que se está analizando.

Otro ejemplo más familiar, es el de la predicción del valor de un bien inmueble. 

Veamos un ejemplo de reconocimiento


```python
import matplotlib.pyplot as plt
import matplotlib.lines as lines
import numpy as np
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
# Suponga que X es el dinero invertido
# e y es el dinero ganado
X = np.array([0,6.5,13.2,18.1,24.9,28.2,36.1])
y = np.array ([5, 10,  17,  20,  25,  30,  35])
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(7,7))
plt.scatter(X, y, s=60)
plt.ylim(bottom=0)
plt.title("Valores")
plt.xlabel("Dinero Invertido")
plt.ylabel("Dinero Ganado")
plt.grid()
plt.show()
```

### Ensayo y error

```python
# Dados los posibles valores en el eje x (Dinero invertido) que se muestra en la gráfica
# generamos el siguiente arreglo
X2 = np.arange(0,40)

# Digamos que, usando el ensayo y error, dibujamos una línea que se ajuste a los valores de la gráfica
# Probamos con y2 = 0.2 * X2
y2 = 0.2 * X2
```

```python
plt.figure(figsize=(7,7))
plt.plot(X2, y2,color="red",label="Nuestra solución")
plt.scatter(X, y, s=60)
plt.ylim(bottom=0)
plt.title("Valores")
plt.xlabel("Valores X")
plt.ylabel("Valores Y")
plt.grid()
plt.show()
```

```python
# Muy malo el ajuste
# Probamos con y3 = 0.6 * X2
y3 = 0.6 * X2
```

```python
plt.figure(figsize=(7,7))
plt.plot(X2, y3,color="red",label="Nuestra solución")
plt.scatter(X, y, s=60)
plt.ylim(bottom=0)
plt.title("Valores")
plt.xlabel("Valores X")
plt.ylabel("Valores Y")
plt.grid()
plt.show()
```

```python
# Sigue siendo malo el ajuste
# Probamos con y4 = 0.8 * X2
y4 = 0.8 * X2
```

```python
plt.figure(figsize=(7,7))
plt.plot(X2, y4,color="red",label="Nuestra solución")
plt.scatter(X, y, s=60)
plt.ylim(bottom=0)
plt.title("Valores")
plt.xlabel("Valores X")
plt.ylabel("Valores Y")
plt.grid()
plt.show()
```

```python
# Es bueno el ajuste, pero hay un problema, para x = 0, debe dar y = 5
# ¿Qué pasa si incrementamos de 0.8 a 1.8, por ejemplo?
# Probamos con y5 = 0.8 * X2 + 5
y5 = 0.8 * X2 + 5
```

```python
plt.figure(figsize=(7,7))
plt.plot(X2, y5,color="red",label="Nuestra solución")
plt.scatter(X, y, s=60)
plt.ylim(bottom=0)
plt.title("Valores")
plt.xlabel("Valores X")
plt.ylabel("Valores Y")
plt.grid()
plt.show()
```
