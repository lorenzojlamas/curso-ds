```python
sueldos["Lenguajes de programación"]=[str(c).split(',') for c in sueldos["Lenguajes de programación"]]
```

```python
sueldos["Bases de datos"]=[str(c).split(',') for c in sueldos["Bases de datos"]]
```

```python
sueldos_nuevo=sueldos[sueldos.dtypes[sueldos.dtypes != object].index]
sueldos_nuevo.head()
```

```python
entradas_min_col=10
for c in sueldos.columns:
    if c in sueldos.dtypes[sueldos.dtypes == object]:
        if (c != "Lenguajes de programación") and (c != "Bases de datos"):
            print(c)
            obj_df=sueldos[c].str.get_dummies()
        else:
            print(c)
            obj_df=pd.get_dummies(sueldos[c].apply(pd.Series).stack()).sum(level=0)
        keep=obj_df.sum(axis=0)[obj_df.sum(axis=0)>1].index
        obj_df=obj_df[keep]
        sueldos_nuevo=pd.concat([sueldos_nuevo,obj_df],axis=1)
```
