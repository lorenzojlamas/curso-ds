```python
# Queremos escribir una red neural que sea capaz de observar ciertos datos de
# una función f y devolver el valor de y
def f(x):
    y = 3 * x + 1
    return y
```

```python
# Importamos las librerías necesarias
import tensorflow as tf
import numpy as np
from tensorflow import keras
```

```python
# Definimos el modelo de red neuronal
model = tf.keras.Sequential(
    [keras.layers.Dense(units=1, input_shape=[1])]
)
```

```python
# Compilamos el modelo pasando como parámetros el optimizador y la función de 
# pérdida que queremos utilizar
model.compile(optimizer='sgd', loss='mean_squared_error', metrics=['mae'])
```

```python
# Creamos el set de datos de entrenamiento
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0])
ys = np.array([-2.0, 1.0, 4.0, 7.0, 10.0, 13.0])
```

```python
# Llevamos a cabo el ajuste del modelo
model.fit(xs, ys, epochs=500)
```

```python
# Creamos un set de test
xst = np.array([10.0, 50.0, 100.0, 222.0])
yst = np.array([31.0, 151.0, 301.0, 667.0])
```

```python
# Probamos una única predicción
print(model.predict([10.0]))
```

```python
# Usamos el set de test
ysp = model.predict(xst)
ysp
```

```python
# Evaluamos el modelo
model.evaluate(xst, yst)
```
